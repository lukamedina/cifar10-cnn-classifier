```python
from RedNeuronal.Model import NeuronalNewtork
import torch
from torch import nn
from torchvision import datasets,transforms
from torch.utils.data import dataloader

#Ruta
path_model = "./Modelos/my_modelo.pth"
path_testing = "./testing_set/"

Model = NeuronalNewtork()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

Model.load_state_dict(torch.load(path_model,map_location=device))
Model.to(device)

Model = Model.eval()

opti = torch.optim.Adam(params=Model.parameters(),lr=2e-4,betas=(0.9, 0.999),weight_decay=5e-4)
loss_fn = nn.CrossEntropyLoss()

png_to_tensor = transforms.Compose([
    transforms.Resize((32,32)),
    transforms.ToTensor(),
    transforms.Normalize((.5,.5,.5),(.5,.5,.5)) 
])

data_set = datasets.ImageFolder(path_testing,transform=png_to_tensor)
dl = dataloader.DataLoader(dataset=data_set,shuffle=True,num_workers=6,batch_size=64) 
```

```python
print(data_set.class_to_idx)
```

```python
#testing

correct = 0
total = 0

with torch.no_grad():
    for x, y in dl:
        x = x.to(device)
        y = y.to(device)

        outputs = Model(x)
        _, preds = torch.max(outputs, 1)

        correct += (preds == y).sum().item()
        total += y.size(0)

print(f"Correctos : {100 * correct / total}%") 
    
```
